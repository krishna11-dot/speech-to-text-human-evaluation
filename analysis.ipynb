```python
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
from matplotlib.transforms import Bbox
import colorcet as cc
import numpy as np
import seaborn as sns
import os
```

```python
SOURCES_PATH = "./"
results_file = os.path.join(
    SOURCES_PATH,
    "Comparing_Speech-to-Text_Software.csv"
)

raw_data = pd.read_csv(results_file)
raw_data = raw_data.dropna(subset="Timestamp")
raw_data
```

```python
clean = raw_data.drop(columns=[
    "Timestamp",
    "Consent form for participation in the research project Human Evaluation of Speech-to-Text Systems",
    "What is your name?"])
clean
```

```python
comments = clean.iloc[:,[8,9,16,17]]
comments
```

```python
answers = clean.iloc[:,[0,1,2,3,4,5,6,7,10,11,12,13,14,15]]

def get_test_answers(answers, test_number):
    return answers.iloc[:, test_number+1]

british_answers = answers[answers["What is your Nationality?"] == "British"]
romanian_answers = answers[answers["What is your Nationality?"] == "Romanian"]
italian_answers = answers[answers["What is your Nationality?"] == "Italian"]
indian_answers = answers[answers["What is your Nationality?"] == "Indian"]

get_test_answers(answers, 1)
```

```python
ethnicities = ["British", "Romanian", "Italian", "Indian"]
ethnicities_colours = sns.color_palette(n_colors=len(ethnicities))
ethnicities_colours
```

```python
def plot_testcase_hist(ax, results, title):
    bins = np.arange(9) - 0.5
    ax.hist(results, bins=bins, label=list(range(1,8)), histtype='bar', stacked=True, color=ethnicities_colours)
    ax.set_xlim(0.5, 7.5)
    ax.set_ylim(0, 20)
    # put the legend
    ax.legend(ethnicities, loc="best")
    ax.set_title(title)

def plot_testcase_hists(n_test_case):
    fig, axs = plt.subplots(1, 2, figsize=(20, 5))
    grouped_answers_google = [
        get_test_answers(british_answers, n_test_case),
        get_test_answers(romanian_answers, n_test_case),
        get_test_answers(italian_answers, n_test_case),
        get_test_answers(indian_answers, n_test_case)
    ]
    grouped_answers_microsoft = [
        get_test_answers(british_answers, n_test_case+6),
        get_test_answers(romanian_answers, n_test_case+6),
        get_test_answers(italian_answers, n_test_case+6),
        get_test_answers(indian_answers, n_test_case+6)
    ]
    plot_testcase_hist(axs[0], grouped_answers_google, f"Google: Test Case {n_test_case}")
    plot_testcase_hist(axs[1], grouped_answers_microsoft, f"Microsoft: Test Case {n_test_case}")
    plt.savefig(f"figures/testCase{n_test_case}.pdf")
```

# Nationality Aggregated Comparison

```python
def concatenate_answers(answers, min, max):
    concatenated = None
    for i in range(min,max+1):
        test_answers = get_test_answers(answers, i)
        if concatenated is not None:
            concatenated = pd.concat([concatenated, test_answers])
        else:
            concatenated = test_answers.copy()

    return concatenated

def concatenate_google_answers(answers):
    return concatenate_answers(answers, 1, 6)

def concatenate_microsoft_answers(answers):
    return concatenate_answers(answers, 7, 12)

def plot_aggregated_hist(ax, results, title):
    bins = np.arange(9) - 0.5
    ax.hist(results, bins=bins, label=list(range(1,8)), histtype='bar', stacked=True, color=ethnicities_colours)
    ax.set_xlim(0.5, 7.5)
    ax.set_ylim(0, 70)
    # put the legend
    ax.legend(ethnicities, loc="best")
    ax.set_title(title)

def plot_ethnicity_hists():
    _, axs = plt.subplots(1, 2, figsize=(20, 5))
    grouped_likert_google = [
        concatenate_google_answers(british_answers),
        concatenate_google_answers(romanian_answers),
        concatenate_google_answers(italian_answers),
        concatenate_google_answers(indian_answers)
    ]
    grouped_likert_microsoft = [
        concatenate_microsoft_answers(british_answers),
        concatenate_microsoft_answers(romanian_answers),
        concatenate_microsoft_answers(italian_answers),
        concatenate_microsoft_answers(indian_answers)
    ]
    plot_aggregated_hist(axs[0], grouped_likert_google, f"Google: Aggregated Score")
    plot_aggregated_hist(axs[1], grouped_likert_microsoft, f"Microsoft: Aggregated Score")
    plt.savefig(f"figures/aggregatedScores.pdf")

plot_ethnicity_hists()
```

# Ease of Use

```python
def get_ease_of_use_by_nationality(nationality, results):
    answers_by_nationality = results[results["What is your Nationality?"] == nationality]
    return answers_by_nationality.iloc[:, [8, 16]]


def plot_ease_of_use_hist(ax, results, title):
    bins = np.arange(9) - 0.5
    ax.hist(results, bins=bins, label=list(range(1,8)), histtype='bar', stacked=True, color=ethnicities_colours)
    ax.set_xlim(0.5, 7.5)
    ax.set_ylim(0, 12)
    # put the legend
    ax.legend(ethnicities, loc="best")
    ax.set_title(title)


def plot_ease_of_use(results):
    british = get_ease_of_use_by_nationality("British", results)
    romanian = get_ease_of_use_by_nationality("Romanian", results)
    italian = get_ease_of_use_by_nationality("Italian", results)
    indian = get_ease_of_use_by_nationality("Indian", results)

    google_british, microsoft_british =  british.iloc[:, 0], british.iloc[:, 1]
    google_romanian, microsoft_romanian =  romanian.iloc[:, 0], romanian.iloc[:, 1]
    google_italian, microsoft_italian =  italian.iloc[:, 0], italian.iloc[:, 1]
    google_indian, microsoft_indian =  indian.iloc[:, 0], indian.iloc[:, 1]
    grouped_ease_of_use_google = [
        google_british,
        google_romanian,
        google_italian,
        google_indian
    ]
    grouped_ease_of_use_microsoft = [
        microsoft_british,
        microsoft_romanian,
        microsoft_italian,
        microsoft_indian
    ]

    _, axs = plt.subplots(1, 2, figsize=(20, 5))
    plot_ease_of_use_hist(axs[0], grouped_ease_of_use_google, f"Google: Ease of Use")
    plot_ease_of_use_hist(axs[1], grouped_ease_of_use_microsoft, f"Microsoft: Ease of Use")
    plt.savefig(f"figures/easeOfUse.pdf")

plot_ease_of_use(clean)
```

# Statistical Tests
## Hypotheses
There is no difference in user preference between the two systems.
There is no difference in ease of use between the two systems.
There is no difference in accuracy between the two systems.
User Nationality has no effect on system accuracy.

## There is no difference in user preference between the two systems.

```python
sp.stats.binomtest(19, 20)
```

## There is no difference in ease of use between the two systems.

```python
ease_of_use = clean.iloc[:, [8, 16]]
google_eou = ease_of_use.iloc[:, 0]
microsoft_eou = ease_of_use.iloc[:, 1]

sp.stats.mannwhitneyu(google_eou.to_numpy(), microsoft_eou.to_numpy())
```

## There is no difference in accuracy between the two systems

```python
def get_sentence_pair_results(results, first, second):
    first_loc = get_test_answers(results, first)
    second_loc = get_test_answers(results, second)
    return np.concatenate((first_loc.to_numpy(), second_loc.to_numpy()))


def get_test_case_results(results, difficulty="easy"):
    test_modes = {
        "easy": [1,2],
        "medium": [3,4],
        "hard": [5,6]
    }
    google_first = test_modes[difficulty][0]
    google_second = test_modes[difficulty][1]
    microsoft_first = google_first+6
    microsoft_second = google_second+6

    google_results = get_sentence_pair_results(results, google_first, google_second)
    microsoft_results = get_sentence_pair_results(results, microsoft_first, microsoft_second)
    return google_results, microsoft_results

easy_results = get_test_case_results(answers, "easy")
medium_results = get_test_case_results(answers, "medium")
hard_results = get_test_case_results(answers, "hard")
```

### Easy Sentences

```python
sp.stats.mannwhitneyu(easy_results[0], easy_results[1])
```

### Medium Sentences

```python
sp.stats.mannwhitneyu(medium_results[0], medium_results[1])
```

### Hard Sentences

```python
sp.stats.mannwhitneyu(hard_results[0], hard_results[1])
```

## User Nationality has no effect on system accuracy.

```python
british_overall = concatenate_google_answers(british_answers), concatenate_microsoft_answers(british_answers)
romanian_overall = concatenate_google_answers(romanian_answers), concatenate_microsoft_answers(romanian_answers)
italian_overall = concatenate_google_answers(italian_answers), concatenate_microsoft_answers(italian_answers)
indian_overall = concatenate_google_answers(indian_answers), concatenate_microsoft_answers(indian_answers)

print(f"British Accuracy Difference: {sp.stats.mannwhitneyu(british_overall[0], british_overall[1])}")
print(f"Romanian Accuracy Difference: {sp.stats.mannwhitneyu(romanian_overall[0], romanian_overall[1])}")
print(f"Italian Accuracy Difference: {sp.stats.mannwhitneyu(italian_overall[0], italian_overall[1])}")
print(f"Indian Accuracy Difference: {sp.stats.mannwhitneyu(indian_overall[0], indian_overall[1])}")
```
